```python
sns.distplot(df['Rating'])
plt.axvline(x=np.mean(df['Rating']) ,c='red', ls='--')
plt.axvline(x=np.percentile(df['Rating'], 25) , c='green', ls='--')
```

```python
sns.distplot(df['Rating'])
plt.axvline(x=np.mean(df['Rating']) ,c='red', ls='--')
plt.axvline(x=np.percentile(df['Rating'], 25) , c='green', ls='--')
plt.axvline(x=np.percentile(df['Rating'], 75) , c='blue', ls='--')
```

```python
sns.distplot(df['Rating'])
plt.axvline(x=np.mean(df['Rating']) ,c='red', ls='--', label='mean')
plt.axvline(x=np.percentile(df['Rating'], 25) , c='green', ls='--', label='25th to 75th percentile')
plt.axvline(x=np.percentile(df['Rating'], 75) , c='blue', ls='--', label='75th-100 percentile') 
plt.legend()
```

Bivariate Analysis

Is there a relationship between gross income and customer ratings?


```python
# Assuming df is your dataframe
sns.scatterplot(x='Rating', y='gross income', data=df)
plt.xlabel('Rating')
plt.ylabel('Gross Income')
plt.title('Scatter Plot: Rating vs. Gross Income')
plt.show()
```

```python
sns.scatterplot(df['Rating'],df['gross income'])
```

```python
sns.regplot(df['Rating'],df['gross income'])
```

```python
# Assuming df is your dataframe
sns.regplot(x='Rating', y='gross income', data=df)
plt.xlabel('Rating')
plt.ylabel('Gross Income')
plt.title('Regression Plot: Rating vs. Gross Income')
plt.show()
```
